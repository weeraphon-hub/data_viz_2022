### ตารางข้อมูล covid

```python
import pandas as pd  #ทำงานกับข้อมูลลักษณะตาราง
import os  #จัดการไฟล์และ path
```

```python
from google.colab import drive  #เชื่อม google drive
drive.mount('/content/drive')
path = '/content/drive/My Drive/data confirmed_cases'   #pathจะชี้ไปยังโฟล์เดอร์
```

```python
data_file_path = os.path.join(path,'confirmed-cases.xlsx') #แล้วก็ชี้ไปที่ไฟล์
print(data_file_path)
```

```python
data_covid = pd.read_excel(data_file_path)  #กำหนดตัวแปร data_covid เป็นข้อมูลของผู้ป่วยโควิด
data_covid  #เรียกดูตารางของผู้ป่วยโควิด
```

```python
data_covid_seek = data_covid[['sex','age','risk']] #แสดงเฉพาะข้อมูล sex,age,risk
data_covid_seek #เรียกดูตาราง
```

### class เป็น risks (เลือกมาแค่ 3 risks)

```python
set(data_covid_seek['risk']) #แปลงค่าที่มีใน column แค่ อย่างละ 1 ตัว
```

####เลือก risk ที่ต้องการ

```python
risk1 = data_covid_seek[data_covid_seek['risk']=='Cluster The Lounge Salaya'] #เลือกเอาเฉพาะ risk ที่ต้องการ
risk1
```

```python
risk2 = data_covid_seek[data_covid_seek['risk']=='Cluster Icon siam'] #เลือกเอาเฉพาะ risk ที่ต้องการ
risk2
```

```python
risk3 = data_covid_seek[data_covid_seek['risk']=='Sandbox'] #เลือกเอาเฉพาะ risk ที่ต้องการ
risk3
```

#### รวมตาราง risk

```python
three_risk = pd.concat([risk1,risk2,risk3]) #รวมตารางของแต่ละ risk
three_risk
```

#### ตรวจค่า missing

```python
three_risk.isnull().any()  #ดูค่า Missing รายคอลัมน์ของตาราง data_covid
#เนื่องจากมี True แสดงว่ามีค่าที่เป็น Missing จะทำการลบค่า Missing ต่อไป
```

### วาดกราฟ parallel coordinate

```python
n1= three_risk.replace('ชาย',0) #แทนชาย = 0
n1
```

```python
n2= n1.replace('หญิง',100) # แทนหญิง = 0
n2
```

```python
compare_risk = n1.replace('หญิง',100) # เปลี่ยนชื่อตัวแปรเป็น compare_risk เพื่อความเรียบร้อย
compare_risk
```

```python
compare_risk['risk'].unique() #เพื่อดู risk ที่เราเลือก
```

```python
pd.plotting.parallel_coordinates(compare_risk,'risk'); # parallel coordinates : คือเอาค่าแต่ละคอลัมน์มาพล็อตเทียบกัน โดยคอลัมน์ risk จะแบ่งสีแบ่งเส้น
```

```python
pd.plotting.parallel_coordinates(compare_risk, 'risk',color=['#33FF99','#CC0066','#FFFF00']); #HTML color บอกค่าของ r g b แค่เป็นเลขฐาน 16
```
